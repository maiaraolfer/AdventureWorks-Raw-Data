```python
# libraries
import pandas as pd
import numpy as np
```

```python
# # TODO 1:load all the datasets
df2020 = pd.read_csv('/Users/maiara/Documents/AdventureWorks Raw Data/AdventureWorks Sales Data 2020.csv')
df2021 = pd.read_csv('/Users/maiara/Documents/AdventureWorks Raw Data/AdventureWorks Sales Data 2021.csv')
df2022 = pd.read_csv('/Users/maiara/Documents/AdventureWorks Raw Data/AdventureWorks Sales Data 2022.csv')


```

```python
# Data exploration

# TODO 2: check the number of rows and columns, data types of each column, missing values.

list = [df2020,df2021,df2022]
for df in list:
    print(df.head()) #mostra as 5 primeiras linhas
    print(df.tail()) #mostra as 5 ultimas linhas
    print(df.info()) #info on all column names and their indexes, if there are null cells, dtype of each column
    print(df.describe()) #count, mean, standar deviation, minimum and maximum values, first, second and third quantiles.
```

```python
# # TODO 3: combine the three datasets into a single dataframe using pd.concat() function.

# Documentation: pd.concat() is a function in pandas library to concatenate objects (series, dataframes etc)
# pd.concat(objs, axis=0, join='outer', ignore_index=False, keys=None, names=None, verify_integrity=False, sort=False, copy=None)
# where objs is a sequence or mapping for pandas objects to concatenate
# axis = 0 sets the axis to concatenate along: axis = 0 (default) - concatenate vertically, stacking objects on top of each other
# axis = 1 - concatenates horizontally, column-wise

list = [df2020,df2021,df2022]
df = pd.concat(list, axis=0) #in this case we want to concatenate the dataframes vertically
print(df.head())
print(df.tail())
```

```python
# How about if we add the product price to the dataframe? 
product_lookup = pd.read_csv('/Users/maiara/Documents/AdventureWorks Raw Data/AdventureWorks Product Lookup.csv')

df = df.merge(product_lookup[['ProductKey', 'ProductPrice', 'ProductCost']], on = "ProductKey", how = 'left')
print(df.head())
print(df.info())
print(df.describe())
# product_lookup[['ProductKey', 'ProductPrice']] tells pandas to only bring in those two columns from the second DataFrame.
# the use of two brackets [[]] is because we are getting more than one column from product_lookup: 
# one pair of brackets > pandas will handle as a series; two pairs > handles as dataframe
# on='ProductKey' tells pandas to match rows using the ProductKey column.
# how='left' keeps all rows from df2020 (sales) even if a matching price is not found.
```

Now, lets add a new column to calculate something, like the profit or discount using Numpy.

```python
# TODO 4: Calculate the revenue and add it as new column.

df['Revenue'] = df['ProductPrice'] * df['OrderQuantity']
print(df.head())                     
```

```python
# TODO 5: Calculate average Revenue for each product over the years and filter revenue higher than 1000.

# Get unique ProductKeys
product_keys = df['ProductKey'].unique()
avg_revenue_list = []
# Loop through each ProductKey and calculate the mean
for key in product_keys:
    avg_revenue = np.mean(df[df['ProductKey'] == key]['Revenue'])
    avg_revenue_list.append({'ProductKey': key, 'AverageRevenue':avg_revenue}) #.2f shows 2 digit after decimal point
    
avg_revenue_df = pd.DataFrame(avg_revenue_list)
print(avg_revenue_df)
```

```python
# TODO 6: Filter the average revenue that are higher than some value.

filtered_df = avg_revenue_df[avg_revenue_df['AverageRevenue']>2500]
print(filtered_df)
```

```python
product_keys = df['ProductKey'].unique()
mean_revenue_list = []

for key in product_keys:
    subset = df[df['ProductKey'] == key]
    total_revenue = subset['Revenue'].sum()
    total_quantity = subset['OrderQuantity'].sum()
    mean_revenue = total_revenue / total_quantity if total_quantity != 0 else 0
    mean_revenue_list.append({'ProductKey': key, 
                              'MeanRevenue': mean_revenue,
                              'TotalOrderQuantity': total_quantity})

result_df = pd.DataFrame(mean_revenue_list)
print(result_df)
```

```python
filter_df = result_df[result_df['MeanRevenue']>2500]
print(filter_df)
```
